```python
inventory = {
    "Laptop": 15,
    "Headphones": 8,
    "Keyboard": 20,
    "Mouse": 5,
    "Monitor": 12
}

print("Original Inventory:", inventory)

inventory["Flash Disk"] = 25
inventory["Mouse"] = 10
print("\nUpdated Inventory:", inventory)

def low_stock(inv):
    return {product: qty for product, qty in inv.items() if qty < 10}

print("\nLow Stock Products:", low_stock(inventory))

del inventory["Keyboard"]
print("\nAfter Deleting 'Keyboard':", inventory)

print("\nInventory Items:")
for product, qty in inventory.items():
    print(f"{product}: {qty}")
```
