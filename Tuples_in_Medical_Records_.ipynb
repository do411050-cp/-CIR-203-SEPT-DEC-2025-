```python
from collections import namedtuple

Patient = namedtuple("Patient", ["full_name", "age", "bp", "hr"])

p1 = Patient("Alice Mwangi", 36, "118/78", 74)
print("Patient:", p1)

print("\nAge:", p1.age)
print("Heart Rate:", p1.hr)

print("""
Tuples/NamedTuples are immutable, making them ideal for medical data
since vitals should not change accidentally.
""")

temp_list = list(p1)
temp_list[3] = 80
p1 = Patient(*temp_list)
print("\nUpdated Patient:", p1)

patients = (
    Patient("Brian Odhiambo", 45, "121/80", 79),
    Patient("Sarah Kendi", 29, "115/70", 73),
    Patient("James Kairu", 54, "130/86", 81),
    Patient("Lucy Wairimu", 40, "117/75", 76),
    Patient("Kevin Mulwa", 33, "122/79", 72)
)

names = [pt.full_name for pt in patients]
print("\nAll Patient Names:", names)
```
