```python
class RouteManager:
    def __init__(self):
        self.routes = [
            "Kisii-Suneka", "Nairobi-Kilimani", "Mombasa-Bamburi",
            "Eldoret-Huruma", "Kericho-Kipkelion", "Thika-Kariobangi",
            "Kitale-Kapenguria", "Nakuru-Section58", "Nyamira-Kebirigo",
            "Malindi-Watamu"
        ]

    def update_routes(self):
        self.routes += ["Garissa-Tana"]
        if "Thika-Kariobangi" in self.routes:
            self.routes.pop(self.routes.index("Thika-Kariobangi"))

    def sort_and_reverse(self):
        self.routes = sorted(self.routes)[::-1]

    def count_n(self):
        return len([r for r in self.routes if r.startswith("N")])

    def long_routes(self):
        return [r for r in self.routes if len(r) > 10]


manager = RouteManager()
manager.update_routes()
manager.sort_and_reverse()

print("Updated Routes:", manager.routes)
print("Routes Starting With 'N':", manager.count_n())
print("Long Routes (>10 chars):", manager.long_routes())
```
