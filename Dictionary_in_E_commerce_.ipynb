```python
class Inventory:
    def __init__(self):
        self.stock = dict(
            Phone=12, Charger=4, Earbuds=18, Powerbank=7, Smartwatch=9
        )

    def add_and_update(self):
        self.stock["USB_Cable"] = 22
        self.stock["Earbuds"] = 25  # update

    def low_items(self):
        return {k: v for k, v in self.stock.items() if v < 10}

    def remove_item(self, item):
        if item in self.stock:
            del self.stock[item]

    def display(self):
        for product, qty in self.stock.items():
            print(f"{product}: {qty}")


inv = Inventory()
inv.add_and_update()
print("Low Stock:", inv.low_items())
inv.remove_item("Powerbank")
inv.display()
```
