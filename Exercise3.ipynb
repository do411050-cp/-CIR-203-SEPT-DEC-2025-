```python
patient = ("John Doe", 45, "120/80", 78)
print("Patient:", patient)

print("\nAge:", patient[1])
print("Heart Rate:", patient[3])

print("\nTuples are immutable, which prevents accidental changes to vital records.")

patient_list = list(patient)
patient_list[3] = 82
patient = tuple(patient_list)
print("\nUpdated Patient Tuple:", patient)

patients = (
    ("John Achupe", 35, "120/80", 87),
    ("Mary Jenga", 40, "110/70", 67),
    ("Kevin Otieno", 50, "130/85", 80),
    ("Njeri Wanjiku", 30, "115/75", 82),
    ("Peter Njoroge", 45, "118/78", 74)
)

names = [p[0] for p in patients]
print("\nPatient Names:", names)
```
